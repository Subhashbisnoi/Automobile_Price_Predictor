```python
import pandas as pd
import numpy as np
```

```python
car=pd.read_csv('quikr_car.csv')
```

```python
car.head()
```

```python
car['Price'].unique()
```

```python
#need to address
#year has many non-year values
#price has ask for price
#price obeject to int
#kms_driven has kms with integers
#kms_driven object to int
#kms_driven has nan values
#fule type has nan values
#name is very inconsistant so keep first 3 word of name
```

```python
#cleaning
backup=car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

```python
car
```

```python
car=car[car['Price']!='Ask For Price']
```

```python
car
```

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
car
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car
```

```python
car['fuel_type'].unique()
```

```python
car[car['fuel_type']=='LPG']
```

```python
car.dropna(inplace=True)
```

```python
car
```

```python
car['fuel_type'].unique()
```

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car
```

```python
car=car.reset_index(drop=True)
```

```python
car.info()
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

```python
car
```

```python
car.to_csv('Cleaned_car.csv')
```

```python
#Model
X=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
columns_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                       remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(columns_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]

for i in range (1000):
    
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(columns_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores .append(r2_score(y_test,y_pred))
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(columns_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
scores .append(r2_score(y_test,y_pred))
r2_score(y_test,y_pred)
```
